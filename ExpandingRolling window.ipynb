```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
import yfinance as yf
import seaborn as sns
```

```python
start = datetime.datetime(2020,11,1)
end = datetime.datetime(2023,12,8)
assets = ['BTC-USD']
MSTR=yf.download(assets, start = start, end = end)
```

```python
MSTR['Close'].plot(figsize = (14,5))
```

### Rolling

Rolling with windows size of every 7 days. I will get table, where values are equal to 7 days average. 

So first 7 days are null, because u can not calculate avg. Actual day = mean(last 7 days)

```python
MSTR.rolling(7).mean().head(10)
```

```python
MSTR['Open'].plot()
MSTR.rolling(7).mean()['Close'].plot(figsize=(15,5))
```

```python
MSTR['Open'].plot()
MSTR.rolling(window=30).mean()['Close'].plot(figsize=(15,5))
```

### Expanding window

It is window, where each observation is equal to average of all previous observetions. Meanwhile in rolling window, the window is still moving with the parametr of days, here I take always all values

```python
MSTR['Open'].plot()
MSTR['Close'].expanding(min_periods=1).mean().plot(figsize=(16,5))
```
