```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/Uzivatel/Downloads/tryauta.csv', index_col=[0])
df.head()
```

```python
fix, ax = plt.subplots(nrows=2,ncols=2, figsize=(20,20))
sns.histplot(data = df, x = 'price', kde = True, ax=ax[0,0], bins= 25)
sns.histplot(data = df, x = 'price', kde = True, hue = 'trans', ax=ax[0,1], bins= 25)
sns.histplot(data = df, x='price', kde= True, hue = 'fuel', ax= ax[1,0], bins= 25)
sns.histplot(data = df, x='price', kde= True, hue = 'kraj', ax= ax[1,1], bins= 25)
# Pravá šikmost - napravo od průměru jsou odlehlejší hodnoty
# distribuce ceny je značne odlišna mezi automatem a manualem. Automat je blížko k normálnímu rozdělení

print('avg price for Diesel', df[df['trans'] == 'Automat'].price.mean())
print('avg price fo Benzin', df[df['trans'] == 'Manuál'].price.mean())
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x = 'fuel', data = df)
#sns.countplot(data = df, x='trans')

print('avg price for Diesel', df[df['fuel'] == 'Diesel'].price.mean())
print('avg price fo Benzin', df[df['fuel'] == 'Benzín'].price.mean())
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(data = df, x = 'trans', y='price')
```

```python
fix, ax = plt.subplots(nrows=2,ncols=2, figsize=(20,10))

df_year = df.groupby('year')['price'].mean().reset_index()
df_kw = df.groupby('kw')['price'].mean().reset_index()

sns.scatterplot(data = df, x = 'kw', y = 'price', ax = ax[0,0])
sns.scatterplot(data = df, x = 'year', y = 'price', ax = ax[0,1])
sns.scatterplot(data = df_year, x = 'year', y = 'price', ax = ax[1,1])
sns.scatterplot(data = df_kw, x = 'kw', y = 'price', ax = ax[1,0])
```

```python
fix, ax = plt.subplots(nrows=2,ncols=2, figsize=(20,10))

df_kraj = df.groupby('kraj')['price'].mean().reset_index()
df_km = df.groupby('km')['price'].mean().reset_index()

sns.scatterplot(data = df, x='kraj', y ='price', ax = ax[0,0])
sns.scatterplot(data = df, x='km', y ='price', ax = ax[0,1])
sns.scatterplot(data = df_kraj, x='kraj', y ='price', ax = ax[1,0])
sns.scatterplot(data = df_km, x='km', y ='price', ax = ax[1,1])
```
